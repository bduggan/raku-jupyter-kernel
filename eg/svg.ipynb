# Two examples from SVG::Plot

## Example 1
stdout is svg, result is text/plain.

```raku
use SVG;
use SVG::Plot;

my $points = 50;
my @x  = (0..$points).map: { sin(2 * pi * $_ / $points) };
my @d1 = (0..$points).map: { 2 * cos(2 * pi * $_ / $points) };
my @d2 = (0..$points).map: { cos(2 * pi * $_ / $points) };
say SVG.serialize: SVG::Plot.new(
    width => 400,
    height => 250,
    :@x,
    values => (@d1, @d2),
    title  => 'sin(x/10), cos(x/10)',
).plot(:xy-lines);
12;
```

# Example 2
Result is svg, stdout is text/plain.

```raku
use SVG;
use SVG::Plot;

# cw: Sensible storage method
my @data1 = (
	# [x, y, mag]
	[  1,    5,    1],
	[  3,    6,    1],
	[  5,    4,    1],
	[  2,    2,    1],
);

my @data2 = (
	[ 2,   5, 0.5],
	[ 4,   4, 0.5],
	[ 3, 2.5, 0.5],
	[ 5,   4, 0.5],
);

my $svg = SVG::Plot.new(
    width   		=> 500,
    height  		=> 300,
    values  		=> [ $(@data1), $(@data2) ],
    background		=> 'gray',
    title   		=> 'Bubbles!',
    min-y-axis 		=> 0
).plot(:bubbles, :opacity(0.8));
say "this is plain text";
SVG.serialize($svg);
```
